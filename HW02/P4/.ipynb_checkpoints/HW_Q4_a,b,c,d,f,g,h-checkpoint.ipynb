```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([0,2,0,3,1,2,2,2,0,2,3,1,0,0,2,3,3,2,3,0])

#bar plot
val, cnt = np.unique(df, return_counts=True)
plt.bar(val, cnt , width = 0.3 , color=['black', 'red', 'blue', 'purple'])
plt.xlabel('value')
plt.ylabel('number of occurance')
```

```python
#calculate pmf
def pmf(data):
    val , cnt = np.unique(data , return_counts=True)
    cnt = cnt / max(cnt)
    
    #calculate total probability
    total = 0
    for i in range(len(cnt)):
        total += cnt[i]
    prob = cnt/total
    plt.bar(val,prob, width = 0.3)
    plt.xlabel('value')
    plt.ylabel('probability of each event')
    print('The probability of each number of success' + str(prob))
    print('The normalize occurance of each value' + str(cnt))

pmf(df)
```
