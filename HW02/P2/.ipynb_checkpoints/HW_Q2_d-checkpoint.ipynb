```python
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def get_photo(i):

    images = glob.glob (r"C:\\Users\\Fateme\\Desktop\\inputs\\P2\\dataset\\c\\Train\\/" + str(i) + '*')


    image_data = []
    for img in images:
        this_image = cv2.imread(img)
        image_data.append(this_image)

    avg_image = image_data[0]
    for i in range(len(image_data)):
        if i == 0:
            pass
        else:
            alpha = 1.0/(i + 1)
            beta = 1.0 - alpha
            avg_image = cv2.addWeighted(image_data[i], alpha, avg_image, beta, 0.0)

    
    plt.imshow(avg_image)
    plt.show()
    PHOTO_ARR = np.array(avg_image)
    return PHOTO_ARR
```

```python

def distance(a,b):
    temp =abs(np.subtract(a,b))
    return temp.mean()
```

```python
images = glob.glob (r"C:\\Users\\Fateme\\Desktop\\inputs\\P2\\dataset\\c\\Test\\/*")

test_data = []
test_label = []
for img in images:
    this_image = cv2.imread(img)
    test_data.append(this_image)
    
for i in range(0,10):
    test_label.append(0)
for i in range(10,20):
    test_label.append(1) 
for i in range(20,30):
    test_label.append(2)
for i in range(30,40):
    test_label.append(3)
for i in range(40,50):
    test_label.append(4)
for i in range(50,60):
    test_label.append(5) 
for i in range(60,70):
    test_label.append(6)
for i in range(70,80):
    test_label.append(7) 
for i in range(80,90):
    test_label.append(8)
for i in range(90,100):
    test_label.append(9) 
print(len(test_label))    
 
predict =[]
for i in range(len(test_data)):    
    o = distance(test_data[i],eight)
    l = distance(test_data[i],nine)
    a = distance(test_data[i],zero)
    b = distance(test_data[i],one)
    c = distance(test_data[i],two)
    d = distance(test_data[i],tree)
    f = distance(test_data[i],four)
    g = distance(test_data[i],five)
    h = distance(test_data[i],six)
    i = distance(test_data[i],seven)

    if min(a,b,c,d,f,g,h,i,o,l)== a:
        predict.append(0)
    elif min(a,b,c,d,f,g,h,i,o,l)== b:
        predict.append(1)
    elif min(a,b,c,d,f,g,h,i,o,l)== c:
        predict.append(2)
    elif min(a,b,c,d,f,g,h,i,o,l)== d:
        predict.append(3)
    elif min(a,b,c,d,f,g,h,i,o,l)== f:
        predict.append(4)
    elif min(a,b,c,d,f,g,h,i,o,l)== g:
        predict.append(5)
    elif min(a,b,c,d,f,g,h,i,o,l)== h:
        predict.append(6)        
    elif min(a,b,c,d,f,g,h,i,o,l)== i:
        predict.append(7)
    elif min(a,b,c,d,f,g,h,i,o,l)== o:
        predict.append(8)        
    elif min(a,b,c,d,f,g,h,i,o,l)== l:
        predict.append(9)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(test_label, predict)
score
```
