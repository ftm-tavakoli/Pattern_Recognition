```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')

sns.set_style("whitegrid")
 

a = sns.FacetGrid(iris, hue ="species",
              height = 5).map(plt.scatter,
                              'sepal_length',
                              'sepal_width')
a.fig.suptitle('Distribution of feature 1 and 2')


b = sns.FacetGrid(iris, hue ="species",
              height = 5).map(plt.scatter,
                              'sepal_length',
                              'petal_length')
b.fig.suptitle('Distribution of feature 1 and 3')


c = sns.FacetGrid(iris, hue ="species",
              height = 5).map(plt.scatter,
                              'sepal_width',
                              'petal_width')
c.fig.suptitle('Distribution of feature 2 and 4')
```
